```python
from bs4 import BeautifulSoup 
import requests
```

```python
url = 'https://www.federalreserve.gov/releases/lbr/current/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table=soup.find_all('table')[1]
```

```python
bank_title_header = table.find_all('th')
```

```python
bank_title_header
```

```python
bank_table_title = [title.text.strip() for title in bank_title_header]
print(bank_table_title)
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(columns = bank_table_title)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_bank_row = [data.text.strip() for data in row_data]
    print(individual_bank_row)
```

```python
print(individual_bank_row)
```

```python
for row in column_data[1:]: #Be careful cause empty row may be present, here, at the 1st row it at the very beginning which can cause error
    row_data = row.find_all('td')
    individual_bank_row = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_bank_row
```
